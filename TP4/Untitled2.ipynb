```python
import numpy as np
import matplotlib.pyplot as plt

# Leer el archivo de letras
with open('docs/alphabet_letters.txt') as fp:
    letter_matrices = fp.read().split('\n\n')  # Separar las matrices por doble salto de línea

# Iterar sobre las matrices de letras y generar gráficos para cada una

# Configurar el tamaño de salida y el número de columnas
fig_width = 10  # Ancho de la figura en pulgadas
fig_height = 2  # Alto de la figura en pulgadas
num_columns = 5  # Número de columnas para mostrar las letras

# Calcular el número de filas y el número total de letras
num_letters = len(letter_matrices)
num_rows = num_letters // num_columns  # División entera
if num_letters % num_columns != 0:
    num_rows += 1

# Crear la figura y los ejes
fig, axes = plt.subplots(num_rows, num_columns, figsize=(fig_width, fig_height * num_rows))
fig.subplots_adjust(hspace=0.3)  # Espacio vertical entre subplots

# Iterar sobre las matrices de letras y generar gráficos para cada una
for i, letter_matrix in enumerate(letter_matrices):
    # Crear la matriz a partir del texto de la letra
    rows = letter_matrix.strip().split('\n')
    matrix = np.array([list(map(int, row.split())) for row in rows])

    # Crear una máscara de colores
    colors = np.where(matrix == 1, 'violet', 'white')

    # Obtener las coordenadas del subplot actual
    row_idx = i // num_columns
    col_idx = i % num_columns

    # Crear el gráfico de la matriz con divisiones entre celdas
    ax = axes[row_idx, col_idx]
    ax.axis('off')
    table = ax.table(cellText=None, cellColours=colors, cellLoc='center', loc='center')
    table.scale(1, 1.5)
    table.auto_set_font_size(False)
    table.set_fontsize(14)
    ax.axis('off')

# Ocultar los subplots no utilizados si hay menos de num_rows * num_columns letras
for i in range(num_letters, num_rows * num_columns):
    row_idx = i // num_columns
    col_idx = i % num_columns
    axes[row_idx, col_idx].axis('off')

# Mostrar la figura con los gráficos de las letras
plt.show()

```
